```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
dataset = pd.read_csv('../input/olimpiyat/olimpiyatlar_temizlenmis_20220204.csv')
```

```python
def drawScatter():
    sns.set_style('darkgrid')
    sns.scatterplot(x='boy', y='kilo', data=dataset)
    plt.title('Boy ve Agirlik Iliskisi')


drawScatter()
```
